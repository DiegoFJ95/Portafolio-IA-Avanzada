IMPORTAR EL DATASET

```python
import pandas as pd
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

url = 'https://www.apradie.com/datos/dataus.csv'
response = requests.get(url, headers=headers)
with open('dataus.csv', 'wb') as f:
    f.write(response.content)

# Import as data frames
dataus = pd.read_csv('dataus.csv')

# Display the first few rows of each DataFrame to confirm
print("usfirms DataFrame:")
display(dataus.head())
```

CALCULAR EL RETORNO ANUAL LOGARITMICO A PARTIR DE STOCKPRICE, ORDENANDO LOS VALORES POR FIRMA Y AÑO.

```python
import numpy as np

dataus = dataus.sort_values(by=['firm', 'year'])
dataus['log_return'] = dataus.groupby('firm')['stockprice'].pct_change().apply(lambda x: np.log(1 + x))
display(dataus.head())
```

QUITAR LOS VALORES FALTANTES (POR EJEMPLO EL PRIMER AÑO QUE NO TIENE CAMBIO POR QUE NO HAY UN AÑO ANTERIOR) Y CONVIERTE LA COLUMNA DE INDUSTRIA A CATEGÓRICA.

```python
dataus.dropna(subset=['log_return'], inplace=True)
dataus['industry'] = dataus['industry'].astype('category')
display(dataus.head())
```

CREAR LAS CATEGORÍAS CUYA INTERACCIÓN VAMOS A ESTUDIAR EN EL MODELO, OMITIENDO LA PRIMERA (FOOD AND BEVERAGE) PARA EVITAR LA MULTICOLLINARIEDAD, LUEGO ELIMINA LA COLUMNA ORIGINAL DE INDUSTRIA Y FINALMENTE CREA Y ENTRENA EL MODELO TOMANDO COMO VARIABLES INDEPENDIENTES EL EARNINGS PER SHARE Y LAS CATEGORÁIS DE INDUSTRIA, Y COMO VARIABLE DEPENDIENTE EL RETORNO LOGARÍTMICO ANUAL.

```python
from linearmodels.panel import PanelOLS
import statsmodels.api as sm

dataus = dataus.set_index(['firm', 'year'])

dependent_variable = dataus['log_return']
independent_variables = dataus[['epsp', 'industry']]

# Create interaction terms
for industry_category in dataus['industry'].cat.categories:
    if industry_category != dataus['industry'].cat.categories[0]:  # Avoid multicollinearity by dropping one category
        independent_variables[f'epsp_x_{industry_category}'] = independent_variables['epsp'] * (independent_variables['industry'] == industry_category)

independent_variables = sm.add_constant(independent_variables)

# Drop the original 'industry' column after creating interaction terms
independent_variables = independent_variables.drop('industry', axis=1)

model = PanelOLS(dependent_variable, independent_variables, entity_effects=True)
results = model.fit()

print(results)
```

EL INTERCEPTO DE 0.0589 INDICA EL VALOR BASE CUANDO TODAS LAS VARIABLES INDEPENDIENTES SON 0, Y ES ESTADÍSTICAMENTE SIGNIFICATIVO AL TENER UN T-STAT DE MÁS DE 2 DESVIACIONES ESTÁNDAR (EN ESTE CASO 8.94).

EL VALOR DE EARNINGS PER SHARE TIENE UN EFECTO POSITIVO DE CASI 0.5 EN EL RETORNO LOGARÍTMICO, Y ES ESTADÍSTICAMENTE MUY SIGNIFICATIVO POR SU T-STAT DE 7.15.

PERTENECER A LA INDUSTRIA DE SOFTWARE & DATA TIENE UN EFECTO LIGERAMENTE NEGATIVO DE -0.0773 PUNTOS PORCENTUALES RESPECTO A PERTENECER A FOOD & BEVERAGES PARA EL RETORNO LOGARÍTMICO, PERO NO ES ESTADÍSTICAMENTE SIGNIFICATIVO AL SOLO TENER UN T-STAT DE 1 DISTANCIA DESVIACIÓN ESTÁNDAR, CON UN INTERVALO DE CONFIANZA DE SOLO EL 68.7%.

PERTENECER A LA INDUSTRIA DE TRADE TIENE UN EFECTO NEGATIVO MAYOR RESPECTO A PERTENER A FOOD & BEVERAGES DE -0.3544 PUNTOS PORCENTUALES, Y ESTE EFECTO ES ESTADÍSTICAMENTE SIGNIFICATIVO POR TENER UN T-STAT DE CASI 5 DESVIACIONES ESTÁNDAR, ESTANDO MUY POR ENCIMA DEL UMBRAL DE 95% DE INTERVALO DE CONFIANZA.

EN CONCLUSIÓN EL EFECTO DEL EPSP SÍ CAMBIA DEPENDIENDO LA INDUSTRIA, SIENDO CASI IGUAL ENTRE PERTENECER A SOFTWARE & DATA Y A FOOD & BEVERAGE PERO SIENDO SIGNIFICATIVAMENTE NEGATIVO PARA LA INDUSTRIA DE TRADE.
